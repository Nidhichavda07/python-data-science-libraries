
# NumPy

## Description
Core library for numerical arrays and fast mathematical operations.

## Advantages
- Very fast
- Memory efficient
- Vectorized operations

## Disadvantages
- Only numeric data
- Harder syntax for beginners

## Example Code


```python
import numpy as np

a = np.array([1,2,3,4])
print("Array:", a)
print("Mean:", np.mean(a))
```


# Pandas

## Description
Used for tabular data manipulation with DataFrames.

## Advantages
- Easy CSV/Excel handling
- Powerful filtering/grouping

## Disadvantages
- Slower with huge datasets

## Example Code


```python
import pandas as pd

df = pd.DataFrame({"name":["A","B","C"],"score":[80,90,85]})
print(df)
print("Average:", df["score"].mean())
```


# Statsmodels

## Description
Statistical modeling and regression analysis.

## Advantages
- Detailed statistical results
- Hypothesis testing

## Disadvantages
- Not ML focused

## Example Code


```python
import statsmodels.api as sm
import numpy as np

X = np.arange(10)
y = 2*X + 1

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```


# Polars


# XGBoost / LightGBM

## Description
High-performance gradient boosting for tabular data.

## Advantages
- Very accurate
- Fast training

## Disadvantages
- Hard tuning

## Example Code


```python
from xgboost import XGBClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.2, random_state=42
)

model = XGBClassifier(eval_metric="mlogloss")

model.fit(X_train, y_train)

preds = model.predict(X_test)

print("Test Accuracy:", accuracy_score(y_test, preds))

```


# Dask

## Description
Parallel computing for large datasets.

## Advantages
- Handles big data
- Scales across cores

## Disadvantages
- More complex setup

## Example Code


```python
import dask.dataframe as dd
import pandas as pd

pdf = pd.DataFrame({"a": range(1000)})
ddf = dd.from_pandas(pdf, npartitions=4)

print(ddf.a.mean().compute())
```
